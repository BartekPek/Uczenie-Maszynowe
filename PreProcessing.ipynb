```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'colour': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0 , 89.0, 99.0, 129.0, 79.0],
    'weigh': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}
df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df.info()
```

```python
for col in ['size', 'colour', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weigh'] = df['weigh'].astype('float')
df.info()
```

Standaryzacja - StandardScaler
std() - pandas nieobciążony estymator wariancji

std() - numpy obciążony

```python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
def standarize(x):
  return (x - x.mean()) / x.std()

standarize(df['price'])
```

```python
from sklearn.preprocessing import scale 
scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```

```python
scaler = StandardScaler()
df[['price', 'weigh',]] = scaler.fit_transform(df[['price', 'weigh']])
df
```
